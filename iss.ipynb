```python
import requests
import matplotlib.pyplot as plt
import time

iss_url = "https://api.wheretheiss.at/v1/satellites/25544"

long = []
lat = []
time_atm = []

number = 102 # since ping once every 10 seconds for 17 or so minutes, so (17*60)/10 = 102
for x in range(number):
    where = requests.get(iss_url)
    iss_loc = where.json()
    latitude = iss_loc["latitude"]
    longitude = iss_loc["longitude"]
    timestamp = iss_loc["timestamp"]
    lat.append((latitude))
    long.append((longitude))
    time_atm.append((timestamp))
    time.sleep(10)
    
figure, (plot1, plot2) = plt.subplots(1, 2, figsize = (10, 10))
plot1.plot(time_atm, long, marker = "o", linestyle = "-", color = "r", markersize = 3)
plot1.set_title("Longitude V. Time")
plot1.set_xlabel("Time")
plot1.set_ylabel("Longitude")
plot1.grid(True)
plot2.plot(time_atm, lat, marker = "o", linestyle = "-", color = "r", markersize = 3)
plot2.set_title("Latitude V. Time")
plot2.set_xlabel("Time")
plot2.set_ylabel("Latitude")
plot2.grid(True)
plt.show()
```
